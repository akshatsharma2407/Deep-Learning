```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/cuekoo/Binary-classification-dataset/refs/heads/master/data0/data.csv',header=None)
```

```python
X = data.iloc[:, 1:]
```

```python
Y = data.iloc[:, 0]
```

```python
def perceptron(x,y, lr = 0.1, epoch=1000):

    w1 = w2 = 1
    b = 0

    for _ in range(epoch):

        for i in range(data.shape[0]):

            z = (w1*(x.iloc[i,0])) + (w2*(x.iloc[i,1])) + b

            if z*(y.iloc[i]) < 0:
                w1 = w1 + (lr*y.iloc[i]*x.iloc[i,0])
                w2 = w2 + (lr*y.iloc[i]*x.iloc[i,1])
                b = b + (lr*y.iloc[i])
    
    return w1, w2, b
```

```python
w1, w2, b = perceptron(X,Y)
```
